```python
import torch

# Create a sample tensor with 3 dimensions
tensor_3d = torch.randn(2, 3, 4)
print("Original tensor_3d:")
print(tensor_3d)
print("Original shape:", tensor_3d.shape)
```

```python
tensor_3d.permute(2, 1, 0)
```

```python
# Permute the dimensions: change order from (dim0, dim1, dim2) to (dim2, dim1, dim0)
tensor_permuted = tensor_3d.permute(2, 1, 0)
print("\nTensor after permuting (2, 1, 0):")
print(tensor_permuted)
print("Permuted shape:", tensor_permuted.shape)
```

```python
# Another example: change order from (dim0, dim1, dim2) to (dim1, dim2, dim0)
tensor_permuted_2 = tensor_3d.permute(1, 2, 0)
print("\nTensor after permuting (1, 2, 0):")
print(tensor_permuted_2)
print("Permuted shape:", tensor_permuted_2.shape)
```
